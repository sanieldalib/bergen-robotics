
## 6) Collections (list, tuple, dict, set)

Collections are types that can store multiple basic types in a single variable. You access specific items in a collection by key or index.

| Type | Example | Description |
|------|----------|-------------|
| `list` | `[1, 2, 3]` | Ordered, mutable (editable) collection |
| `tuple` | `(1, 2, 3)` | Ordered, immutable (non-editable) collection |
| `range` | `range(5)` | Sequence of numbers, often used in loops, is exclusive of last number |
| `dict` | `{"name": "RoboOne", "battery": 90}` | Key‚Äìvalue pairs for structured data |
| `set` | `{1, 2, 3}` | Unordered collection of **unique** elements |

````python
list_example = [1, 2, 3]
tuple_example = (1, 2, 3)
range_example = range(1,5)
dict_example = {"name": "RoboOne", "battery": 90}
set_example = {1, 2, 3}

print(list_example, "->", type(list_example).__name__)
print(tuple_example, "->", type(tuple_example).__name__)
print(range_example, [x for x in range_example], "->", type(range_example).__name__)
print(dict_example, "->", type(dict_example).__name__)
print(set_example, "->", type(set_example).__name__)
````

### Lists

Ordered, mutable (editable) collection

````python
num_list = [10, 20, 30]
print("Initial list:", num_list)

# lists are mutable / changeable
num_list.append(40)
print("Added 40 to the list:", num_list)

print("First item:", num_list[0])

# lists are mutable / changeable
num_list[0] = 100
print("Changed first item:", num_list)

print("Item at index 100:", num_list[100]) # this will throw an error
````

### Tuples
Ordered, immutable (non-editable) collection

````python
# tuples are immutable / unchangeable
pt = (3, 4)
print(pt)

x, y = pt
print("x:", x, "y:", y)

pt[0] = 5 # this will throw an error since tuples are immutable
````

### Ranges

## 8) Functions

**Functions** let you capture reusable logic into a named block. Call it once or a hundred times!

### Key Concepts

- **Purpose**: Reduce code duplication, organize logic, and make code readable
- **Parameters**: Inputs that a function accepts (e.g., `radius` in `area_of_circle(radius)`)
- **Return values**: Output the function sends back (use `return` statement)
- **Type hints** (optional): Clarify parameter types and return type ‚Üí `def func(x: int) -> str:`

### When to Use Functions

- **Repeated logic**: Anywhere you write similar code 2+ times
- **Complex tasks**: Break big problems into smaller, testable pieces
- **Team collaboration**: Well-named functions + type hints = clearer intent
- **Sensor/motor control**: Encapsulate robot actions (e.g., `move()`, `read_sensor()`)

### Default Parameters

Parameters can have **default values** ‚Äî if the caller doesn't provide them, the function uses the default.

```python
def greet(name: str, greeting: str = "Hello") -> str:
    return f"{greeting}, {name}!"

print(greet("Alice"))                    # uses default greeting
print(greet("Bob", greeting="Hi"))       # overrides default
```

**Why use defaults?**
- Make functions flexible and optional
- Reduce boilerplate (don't repeat common values)
- Improve API clarity (e.g., most calls use the default)

### Basic Structure

```python
def function_name(param1: type, param2: type = default) -> return_type:
    # code here
    return result
```


````python
# ‚úÖ Example
def area_of_circle(radius: float) -> float:
    PI = 3.14159
    return PI * radius * radius

print(area_of_circle(2.5))
````

## 9) Classes and Inheritance

**Classes** bundle data (attributes) and behavior (methods) into reusable blueprints. Create instances (objects) from them.

**Inheritance** lets you create a new class based on an existing one, reusing and extending functionality. Replacing functionaility is called overriding.

### Key Concepts

- **`class`**: A blueprint/template for creating objects. Defines what attributes and methods instances will have.

- **Constructor (`__init__`)**:  Special method that **automatically runs when you create a new instance**. Use it to set up initial attributes.
  - Example: `Robot("R2D2")` automatically calls `__init__`, initializing the robot's name, position, etc.
  - The name `__init__` tells Python: "This method initializes new objects"

- **`self`**: A variable representing the **current object instance**. Always the first parameter in methods. When you do `robot.move()`, Python passes `robot` as `self` behind the scenes.

- **Attributes**: Data stored on an object (e.g., `self.name`, `self.x`, `self.y`). Each instance has its own copy.

- **Methods**: Functions that belong to a class. They operate on the object's attributes. Methods also have `self` as first parameter.

- **Inheritance (`class Child(Parent)`)**:  Child class **inherits** (gets access to) all attributes and methods from Parent. You can override or extend them.

- **`super()`**: Calls the parent class's method. Useful when overriding‚Äîyou can add new behavior *and* keep the parent's behavior.

### Simple Structure

```python
class Animal:
    def __init__(self, name: str):
        self.name = name
    
    def speak(self) -> str:
        return f"{self.name} makes a sound"

class Dog(Animal):              # Dog inherits from Animal
    def speak(self) -> str:     # override parent's method
        return f"{self.name} barks!"

dog = Dog("Buddy")
print(dog.speak())
```

Below are more detailed examples:


````python
# ‚úÖ Example: Base class
class Robot:
    def __init__(self, name: str):
        self.name = name
        self.x, self.y = 0, 0

    def move(self, dx: int, dy: int):
        self.x += dx
        self.y += dy
        return (self.x, self.y)

    def report(self):
        return f"{self.name} at ({self.x}, {self.y})"

r = Robot("R2")
print(r.report())
print("moved to:", r.move(2, 3))
print(r.report())
````

````python
# ‚úÖ Example: Inheritance
class FlyingRobot(Robot):
    def __init__(self, name: str):
        super().__init__(name)
        self.z = 0

    def ascend(self, dz: int):
        self.z += dz
        return self.z

    def report(self):
        base = super().report()
        return base + f" and altitude {self.z}"

d = FlyingRobot("SkyScout")
d.move(1, 1)
d.ascend(10)
print(d.report())
````
