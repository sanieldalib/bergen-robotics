## 3) Variables and Basic Types in Python

A **variable** is like a labeled box that stores data in memory.  
You assign a value using the `=` sign, and Python remembers it for later.

### Basic types
Basic types, or primatives, hold only one piece of information

| Type | Example | Description |
|------|----------|-------------|
| `int` | `42` | Whole numbers (positive, negative, or zero) |
| `float` | `3.14159` | Numbers with decimals (floating-point) |
| `str` | `"Hello, world"` | Strings ‚Äî sequences of Unicode characters |
| `bool` | `True`, `False` | Logical truth values used in conditions |
| `NoneType` | `None` | Represents ‚Äúno value‚Äù or null |

### Collection types
Collections are types that can store multiple basic types in a single variable. You access specific items in a collection by key or index.

| Type | Example | Description |
|------|----------|-------------|
| `list` | `[1, 2, 3]` | Ordered, mutable (editable) collection |
| `tuple` | `(1, 2, 3)` | Ordered, immutable (non-editable) collection |
| `range` | `range(5)` | Sequence of numbers, often used in loops, is exclusive of last number |
| `dict` | `{"name": "RoboOne", "battery": 90}` | Key‚Äìvalue pairs for structured data |
| `set` | `{1, 2, 3}` | Unordered collection of **unique** elements |



```python
# Basic Types Example
x = 10            # int
pi = 3.14159      # float
name = "RoboOne"  # str
active = True     # bool

print(x, "->", type(x).__name__)
print(pi, "->", type(pi).__name__)
print(name, "->", type(name).__name__)
print(active, "->", type(active).__name__)
```

## 9) Classes and Inheritance


```python
# ‚úÖ Example: Base class
class Robot:
    def __init__(self, name: str):
        self.name = name
        self.x, self.y = 0, 0

    def move(self, dx: int, dy: int):
        self.x += dx
        self.y += dy
        return (self.x, self.y)

    def report(self):
        return f"{self.name} at ({self.x}, {self.y})"

r = Robot("R2")
print(r.report())
print("moved to:", r.move(2, 3))
print(r.report())
```

```python
# ‚úÖ Example: Inheritance
class FlyingRobot(Robot):
    def __init__(self, name: str):
        super().__init__(name)
        self.z = 0

    def ascend(self, dz: int):
        self.z += dz
        return self.z

    def report(self):
        base = super().report()
        return base + f" and altitude {self.z}"

d = FlyingRobot("SkyScout")
d.move(1, 1)
d.ascend(10)
print(d.report())
```
